```python
hi = np.array(sorted(my_img)).reshape(-1, 19)
hi
```

```python
combined_image = []

for row in hi:
    row_images = [np.array(Image.open(img_path)) for img_path in row]
    combined_row = np.hstack(row_images)  
    combined_image.append(combined_row)

final_image = np.vstack(combined_image)

# Visualize the combined image
plt.figure(figsize=(10, 10))
plt.imshow(final_image)
plt.axis('off')
plt.show()
```

```python
original_height = img1.shape[0]
original_width = img1.shape[1]

cropped_image = final_image[:original_height, :original_width]


plt.figure(figsize=(10, 10))
plt.imshow(cropped_image, cmap='gray')  # Assuming the image is grayscale
plt.axis('off')
plt.show()
# Convert the cropped image back to a PIL Image
cropped_pil_image = Image.fromarray(cropped_image)
normalized_cropped_image = (cropped_image * 255).astype(np.uint8)

visualized_pil_image = Image.fromarray(normalized_cropped_image)
visualized_pil_image.save('b5_pred_image_GBIAH_096_visualized.png')
```
